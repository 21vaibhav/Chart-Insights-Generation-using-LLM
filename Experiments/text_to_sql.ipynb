```python
def dataframe_to_schema(df, table_name="survival_titanic"):
    cols = []
    for col, dtype in df.dtypes.items():
        if "int" in str(dtype):
            sql_type = "INT"
        elif "float" in str(dtype):
            sql_type = "FLOAT"
        else:
            sql_type = "TEXT"

        cols.append(f"{col} {sql_type}")

    columns_sql = ",\n    ".join(cols)

    return print(f"""
Table {table_name} (
    {columns_sql}
)
""")

schema = dataframe_to_schema(df)
schema
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM

MODEL_NAME = "defog/sqlcoder-7b-2"
SAVE_DIR = "/content/drive/MyDrive/models/sqlcoder-7b-2"

# Load tokenizer and model
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
model = AutoModelForCausalLM.from_pretrained(
    MODEL_NAME,
    torch_dtype=torch.float16,
    device_map="auto"
)

tokenizer.save_pretrained(SAVE_DIR)
model.save_pretrained(SAVE_DIR)

def generate_sql(user_prompt: str) -> str:
    prompt = f"""
### Question:
{user_prompt}

### SQL Query:
"""

    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=256,
            temperature=0.0,
            do_sample=False,
            pad_token_id=tokenizer.eos_token_id
        )

    result = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return result.split("### SQL Query:")[-1].strip()
# Example
if __name__ == "__main__":
    user_prompt = f"""schema:{schema}

Display survived, pclass, sex,age
Also compute average age of rows where age > 20
Use valid DuckDB SQL"""
    sql = generate_sql(user_prompt)
    print(sql)
```

## Validation fucntion

```python
import sqlglot
from sqlglot import exp

def validate_sql(sql: str, dialect="duckdb") -> bool:
    try:
        sqlglot.parse_one(sql, dialect=dialect)
        return True
    except Exception as e:
        print("❌ SQL validation failed:", e)
        return False
```

## query fixing code

```python
def fix_over_where(sql: str, dialect="duckdb") -> str:
    try:
        tree = sqlglot.parse_one(sql, dialect=dialect)
    except Exception:
        return sql  # Can't parse → return as is

    changed = False

    for node in tree.find_all(exp.Window):
        if node.args.get("where"):
            condition = node.args["where"]
            func = node.this

            # Convert: AVG(x) OVER (WHERE cond)
            # → AVG(x) FILTER (WHERE cond) OVER ()
            new_func = exp.Filter(
                this=func,
                where=condition
            )

            node.set("this", new_func)
            node.set("where", None)
            changed = True

    return tree.sql(dialect=dialect) if changed else sql
```

# Final

```python
def generate_sql(user_prompt: str,max_retires) -> str:
  for _ in range(max_retires):
    prompt = f"""
    ### Question:
    {user_prompt}

    ### valid DuckDB SQL Query:
    """

    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=256,
            temperature=0.0,
            do_sample=False,
            pad_token_id=tokenizer.eos_token_id
        )

    result = tokenizer.decode(outputs[0], skip_special_tokens=True)
    sql = result.split("### valid DuckDB SQL Query:")[-1].strip()
    sql = fix_over_where(sql)
    if validate_sql(sql):
        return sql
  raise RuntimeError("Failed to generate valid SQL")

# Example
if __name__ == "__main__":
    user_prompt = f"""schema:{schema}

Display survived, pclass, sex,age
Also compute average age of rows where age > 20
"""
    sql = generate_sql(user_prompt,max_retires=5)
    print(sql)
```

```python
def execute_llm_sql(con, sql):
    print("🔹 Original SQL:\n", sql)

    # Step 1: validate
    if not validate_sql(sql):
        raise ValueError("Invalid SQL generated by model")

    # Step 2: auto-fix common LLM errors
    fixed_sql = fix_over_where(sql)

    print("🔹 Fixed SQL:\n", fixed_sql)

    # Step 3: final validation
    if not validate_sql(fixed_sql):
        raise ValueError("SQL still invalid after fixing")

    # Step 4: execute
    return con.execute(fixed_sql).df()
```

```python
con = duckdb.connect()
result = execute_llm_sql(con, sql)
print(result)
```
