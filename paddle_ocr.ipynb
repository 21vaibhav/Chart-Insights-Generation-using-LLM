## **Imports**

```python
from paddleocr import ChartParsing
```

## **for time series**

```python

model = ChartParsing(model_name="PP-Chart2Table")
img_path = r"D:\Downloads\timeseries.png"
results = model.predict(input={"image": img_path})
print(results)
```

## **Prin**

```python
for res in results:
    # Extract the table string
    table_str = res['result']
    print("--- Extracted Graph Data ---")
    print(table_str)
```
